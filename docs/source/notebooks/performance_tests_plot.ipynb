```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_context("talk")
```

```python
data = pd.read_csv("performance_tests_results.txt", sep="\t")
```

```python
data
```

```python
sns.set_theme(style="white")
fig = plt.figure(figsize=(12, 6))
grid = sns.lineplot(x="cpus", y="time", data=data, markers=True, marker="o")
grid.set(xlabel="Number of CPUs", ylabel="Time (seconds)", title="Neofox execution time to annotate Balachandran dataset")
grid.annotate(text='16 CPUs, ~27 minutes', xy=(17, 1600), xycoords='data')
grid.annotate(text='32 CPUs, ~15 minutes', xy=(33, 900), xycoords='data')
grid.annotate(text='64 CPUs, ~9 minutes', xy=(65, 580), xycoords='data')
grid.annotate(text='96 CPUs, ~10 minutes', xy=(90, 610), xycoords='data')
sns.despine()
```

```python
fig.savefig("../../figures/neofox_performance_test_balachandran.png")
```
